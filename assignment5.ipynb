```python
import numpy as np
```

```python
import string
```

```python
string_lengths = np.random.randint(low=5,high=10,size=1000)
```

```python
alphabet=(string.ascii_uppercase+string.ascii_lowercase)
```

```python
print(alphabet)
```

```python
no=[]
mst=[]
hash_value=[]
for i in range (0,len(string_lengths)):
    multy=""
    sum=0
    no.append(string_lengths[i])
    for j in range (0,no[0]):
        multy = multy+alphabet[np.random.randint(low=0,high=52,size=1)[0]]
        no.clear()
    for x in range (0,len(multy)):
        sum=sum+ord(multy[x])
    sum=sum*13
    hash_value.append(sum)
    mst.append(multy)
    
```

```python
value_diction=dict(zip(mst,hash_value))
```

```python
value_diction
```

```python
import operator
```

```python
sorted_d = dict(sorted(value_diction.items(), key=operator.itemgetter(1)))
print('Dictionary in ascending order by value : ',sorted_d)
```

```python
sorted_d = dict(sorted(value_diction.items(), key=operator.itemgetter(1),reverse=True))
print('Dictionary in descending order by value : ',sorted_d)
```
