```python
import numpy as np
```

```python
nums = np.random.randint(low=2,high=180,size=1000)
```

```python
nums
```

```python
even=[]
```

```python
prime=[]
```

```python
for num in nums:
    rem=num%2
    if rem==0:
        even.append(num)
    else:
        for divisor in range (2,num):
            rem=num%divisor
            if rem==0:
                break
            else:
                if divisor==num-1:
                    prime.append(num)
D={"even":even,"prime":prime}
```
